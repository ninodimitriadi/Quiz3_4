```python
### გამოიყენეთ მონაცემთა წმედის რომელიმე ფუნქციონალი (მაგ. რომელიმე სვეტის მონაცემები გადაანაწილეთ რამდენიმე ახალ სვეტად, წაშალეთ რომელიმე სვეტი, ცარიელი მნიშვნელობებისთვისგამოიყენეთ სასურველი ფორმატი, ა.შ.) ###
```

```python
#replace ფუნქციით ფუნქციის გამოყენებით Status-ის მნიშვნელობებში Automatic შევცვალე --- გამოსახულებით
df["Status"] = df["Status"].str.replace("Automatic", "---")
df
```

```python
###რიცხვითი ტიპის ველის მონაცემები დაყავით კატეგორიებად cut() ფუნქციის გამოყენებით (მაგ.ასაკის ველი 3 კატეგორიად „Children”, “Adolescent”, “Old”). შესაბამისად დაამატეთ კატეგორიის სვეტი.###
```

```python
# მიწისძვრის მაგნიტუდეი დავყავი კატეგორიებად ძლიერი საშუალო და სუსტი

bins = [1.0, 3.0, 6.0, 10.0]
labels = ["სუსტი", "საშუალო", "ძლიერი"]

df['Magntude strength'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, right=False)
df
```

```python
###ააგეთ თქვენთვის სასურველი 4 განსხვავებული დიაგრამა შემდეგი ტიპებიდან (ხაზოვანი დიაგრამა, სვეტოვანი დიაგრამა, გაფანტულობის დიაგრამა, წრიული დიაგრამა, ჰისტოგრამა).###
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ჰისტოგრამა 
plt.hist(df['Magnitude'], bins=30, color='blue', edgecolor='black')
plt.title('მაგნიტუდები')
plt.xlabel('Magnitude')
plt.ylabel('მნისვნელობა')
plt.show()
```

```python
# ხაზოვანი დიაგრამა
count = df['Depth'].value_counts()

plt.figure(figsize=(10, 6))
count.plot(kind='line', marker='*', color='b')
plt.title('ხაზოვანი დიაგრამა მიწისძვრის სიღრმისთვის')
plt.xlabel('Depth')
plt.ylabel('სიხშირე')
plt.grid(True)
plt.show()
```

```python
# წრიული დიაგრამა
Source_counts = df['Source'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(Source_counts, labels=Source_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('წყარო')
plt.show()
```

```python
#box-plot
plt.boxplot(df['Magnitude'], vert=False)
plt.title('მაგნიტუდების ფლოტი')
plt.xlabel('Magnitude')
plt.show()
```
